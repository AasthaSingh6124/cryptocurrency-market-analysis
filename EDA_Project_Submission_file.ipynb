# Phase 2: Data Cleaning & Manipulation

### STEP 1 — Removing Useless Columns

```python
df.drop(columns=['image', 'max_supply'], inplace=True)
```

- Contains image URLs → no analytical value
- 43.5% missing values → not helpful for analysis. Too incomplete.

### STEP 2 — Handle Missing Values

```python
##Columns with MEDIUM missingness (~32%)
df['price_change_percentage_1y'] = df['price_change_percentage_1y'].fillna(
    df['price_change_percentage_1y'].median())
```

- It’s numeric
- We can't use mean because data not available its better to fill with median.

```python
##Columns with SMALL missingness (< 5%)
small_missing_cols = [
    'price_change_percentage_30d', 'price_change_percentage_1h',
    'market_cap_change_percentage_24h', 'high_24h', 'low_24h',
    'total_supply', 'price_change_percentage_24h', 'price_change_percentage_7d','fully_diluted_valuation','price_change_24h','market_cap_change_24h'                                
]

for col in small_missing_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

- Filled missing values with median

### STEP 3 — Standardize Categorical Columns

```python
cat_cols = ['id', 'symbol', 'name']

for col in cat_cols:
    df[col] = df[col].str.strip().str.lower()
```

- These are identifiers, not categories, but we can still clean formatting(removing unwanted spaces).

### STEP 4 — Convert Date Columns Correctly

```python
date_cols = ['ath_date', 'atl_date', 'last_updated']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.info()
```

### STEP 5 — Remove Duplicates

```python
df.drop_duplicates(inplace=True)            # Removed duplicate values from our dataset
```

# BIVARIATE ANALYSIS

### 1) Numerical vs Numerical (Correlation Analysis)

```python
# Selecting numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
# Correlation matrix
correlation_matrix = df[numeric_cols].corr().round(2)
correlation_matrix
```

-  market_cap shows strong correlation with total_volume (~0.71), meaning large-cap coins typically have high trading activity.
-  market_cap also correlates strongly with fully_diluted_valuation (~0.70).
-  current_price has a perfect correlation with high_24h and low_24h, because these prices reflect the same price movement range.
-  current_price has weak correlation with market_cap, meaning price alone does not determine a coin’s size due to varying circulating supply.

### 2) Categorical vs Numerical (Groupwise Analysis)

```python
# Top 10 coins by market cap
df.groupby('name')['market_cap'].mean().sort_values(ascending=False).head(10)
```

- Bitcoin and Ethereum dominate the market with the highest market caps.
- The top 10 coins contribute a major portion of total market valuation,showing a heavily concentrated crypto market.

### Average current price by symbol:

```python
df.groupby('name')['current_price'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby('symbol')['current_price'].mean().sort_values(ascending=False).tail(10)
```

# PHASE-3 VISUALIZATIONS

### 1) Distribution of Current Prices

```python
plt.figure(figsize=(10,5))
sns.histplot(df['current_price'], bins=100, kde=True)
plt.title("Distribution of Current Cryptocurrency Prices")
plt.xlabel("Current Price")
plt.ylabel("Frequency")
plt.xscale("log")        # log scale for better visibility
plt.show()
```

- Cryptocurrency prices vary widely. Most coins have very low prices, but a few top cryptocurrencies have extremely high values. This creates a long-tail distribution, which is why the plot looks skewed to the right.

### 2) Market Cap Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(df['market_cap'], bins=50, kde=True,color='red')
plt.title("Distribution of Market Capitalization")
plt.xlabel("Market Cap")
plt.ylabel("Frequency")
plt.show()
```

- Market cap values are extremely skewed.
- Only a few cryptocurrencies dominate the market (Bitcoin, Ethereum), while most coins are small-cap.

### 3) Market Cap vs Total Volume (Scatter Plot)

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='market_cap', y='total_volume',color='green')
plt.xscale("log")
plt.yscale("log")
plt.title("Market Cap vs Total Volume")
plt.xlabel("Market Cap")
plt.ylabel("Total Trading Volume")
plt.show()
```

- There is a positive relationship between market cap and trading volume.
- Larger cryptocurrencies generally have higher trading activity.

### 4. Top 10 Cryptocurrencies by Market Cap (Bar Plot)

```python
top10 = df.nlargest(10, 'market_cap')
plt.figure(figsize=(12,6))
sns.barplot(data=top10, x='name', y='market_cap',color='purple')
plt.title("Top 10 Cryptocurrencies by Market Capitalization")
plt.ylabel("Market Cap")
plt.xticks(rotation=45)
plt.show()
```

- Bitcoin and Ethereum dominate the crypto market.
- Other coins have significantly smaller market caps in comparison.

### 5) Correlation Heatmap of Numerical Variables

```python
plt.figure(figsize=(14,10))
sns.heatmap(df[numeric_cols].corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

- Market Cap & Total Volume

### 6) Price vs Market Cap (Scatter Plot)

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='current_price', y='market_cap')
plt.xscale("log")
plt.yscale("log")
plt.title("Price vs Market Cap")
plt.xlabel("Current Price")
plt.ylabel("Market Cap")
plt.show()
```
